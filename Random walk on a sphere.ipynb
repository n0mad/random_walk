```python
%matplotlib inline
import numpy as np
import pylab as P

# L_2 normalization: x <- x / ||x||
def normalize(x):
    return x / np.sqrt( (x * x).sum() )
```

```python
def tangent_plane_basis(x):
    """Returns a basis the plane that is tangent to the unit sphere at the x"""
    n = x.shape[0]
    #as a simple approach to get an invertable matrix with x as a column is just a replace a column in a diagonal matrix with x
    #since it is unlikely that x will be collinear to another column. However, to get a better stability it might be better
    #to find the component of x with the highest absolute value and put x in the corresponding column of eye(n)
    A = np.eye(n)
    A[:, 0] = x
    Q, R = np.linalg.qr(A)
    basis = Q[:, 1:]
    return basis
```

```python
def random_step(x, e, basis):
    center = np.zeros(basis.shape[1])
    step = np.random.normal(center, e)
    d = np.dot(basis, step)
    if len(d.shape) > 1:
        d = d[:, 0]

    x_new = x + d
    x_new = normalize(x_new)
    assert np.allclose( np.dot(x, d), 0) #just checking that d lies in the tangent plane, so it is orthogonal to x
    return x_new
```

```python
def do_walk(n_dim, n_steps, e):
    """Performs n_steps of random walk on a n_dim-dimensional unit sphere. The steps are performed by sampling an 
    normal distributed vector in a tangent plane. The sampled vector has a variance of e."""
    trace = []

    x_init = np.random.normal( np.zeros(n_dim), 1)
    x_init = normalize(x_init)

    trace.append(x_init)

    for i in xrange(n_steps):
        basis = tangent_plane_basis(trace[-1])
        x_new = random_step(trace[-1], e, basis)
        trace.append(x_new)

    trace = np.array(trace)    
    return trace
```

```python
def plot():
    trace = do_walk(3, 3000, 1)
   
    x = trace[:, 0]
    y = trace[:, 1]

    symmetry_check = []
    symmetry_check.append(((x > 0) * (y > 0)).sum())
    symmetry_check.append(((x < 0) * (y > 0)).sum())
    symmetry_check.append(((x < 0) * (y < 0)).sum())
    symmetry_check.append(((x > 0) * (y < 0)).sum())

    print 'four quarters of the sphere should have approx. equal number of points:'
    print symmetry_check
    
    P.scatter(x, y) 
    P.show()

```

```python
plot()
```

```python
def sample_sphere(n_samples, dim):
    """Sampling n_samples points from the uniform distribution over dim-dimensional sphere""" 
    sample = []
    center = np.zeros(dim)
    for _ in xrange(n_samples):
        sample.append(normalize(np.random.normal(center, 1)))
    return np.array(sample)
```

```python

sample = sample_sphere(3000, 3)
x = sample[:, 0]
y = sample[:, 1]


P.scatter(x, y) 
P.show()
```
